5. Assign students into 4 performance groups using NTILE().

```python
To assign students into 4 performance groups using the SQL NTILE() function, you can rank them by their score and divide them evenly into quartiles.

Here’s how the query would look:

SELECT
    student_id,
    name,
    course,
    score,
    NTILE(4) OVER (ORDER BY score DESC) AS performance_group
FROM Student_Performance;

Result (based on the given data)

| Student Name | Score | Performance Group |
| --- | --- | --- |
| **Arjun Verma** | 95 | 1 (Top Performers) |
| **Tanya Desai** | 93 | 1 |
| **Siddharth Jain** | 92 | 1 |
| **Kabir Khanna** | 91 | 1 |
| **Rohan Gupta** | 89 | 2 |
| **Aarav Mehta** | 88 | 2 |
| **Tanvi Patel** | 84 | 2 |
| **Meera Pillai** | 82 | 2 |
| **Nikhil Rao** | 79 | 3 |
| **Riya Singh** | 76 | 3 |
| **Sneha Kulkarni** | 74 | 3 |
| **Dev Sharma** | 73 | 3 |
| **Vuyra Rao** | 71 | 4 |
| **Ayesha Khan** | 67 | 4 (Needs Improvement) |

Interpretation.
. Group 1: Exceptional performers (scores ≥ 91)
. Group 2: Strong performers (scores 82–89)
. Group 3: Average performers (scores 73–79)
. Group 4: Below-average performers (scores ≤ 71)
```

6.  For each course, assign a row number based on attendance (highest first).

```python
To assign a row number based on attendance (highest first) for each course, you can use the SQL ROW_NUMBER() window function with a partition by course.

Here's the query.
SELECT
    student_id,
    name,
    course,
    attendance,
    ROW_NUMBER() OVER (PARTITION BY course ORDER BY attendance DESC) AS attendance_rank
FROM Student_Performance;

Result (based on your dataset)

| Course | Student Name | Attendance | Attendance Rank |
| --- | --- | --- | --- |
| **Data Science** | Ananya Verma | 96 | 1 |
|  | Aarav Mehta | 92 | 2 |
|  | Riya Singh | 85 | 3 |
|  | Nikhil Rao | 82 | 4 |
| **Python** | Kabir Khanna | 96 | 1 |
|  | Siddharth Jain | 90 | 2 |
|  | Yuvraj Rao | 84 | 3 |
|  | Ayesha Khan | 81 | 4 |
| **SQL** | Sia Bansal | 94 | 1 |
|  | Rohan Gupta | 91 | 2 |
|  | Tanvi Patel | 89 | 3 |
|  | Dev Sharma | 78 | 4 |
| **Tableau** | Arjun Verma | 98 | 1 |
|  | Meera Pillai | 87 | 2 |
|  | Sneha Kulkarni | 83 | 3 |

insight.
. Arjun Verma (Tableau) has the highest attendance overall (98%).
. Dev Sharma (SQL) has the lowest attendance (78%).
. Attendance is generally strong across all courses, with most students above 80%.
```

10. For each course, find the highest score using FIRST_VALUE().

```python
To find the highest score for each course using the SQL FIRST_VALUE() window function, you can partition the data by course and order by score in descending order.

SELECT
    student_id,
    name,
    course,
    score,
    FIRST_VALUE(score) OVER (PARTITION BY course ORDER BY score DESC) AS highest_score
FROM Student_Performance;

Result (based on your dataset)

| Course | Student Name | Score | Highest Score |
| --- | --- | --- | --- |
| **Data Science** | Tara Mehta | 93 | 93 |
|  | Aarav Mehta | 88 | 93 |
|  | Riya Singh | 76 | 93 |
|  | Nikhil Rao | 79 | 93 |
| **Python** | Kabir Khanna | 91 | 91 |
|  | Siddharth Jain | 85 | 91 |
|  | Yuvraj Rao | 71 | 91 |
|  | Ayesha Khan | 67 | 91 |
| **SQL** | Sia Bansal | 92 | 92 |
|  | Rohan Gupta | 88 | 92 |
|  | Tanvi Patel | 84 | 92 |
|  | Dev Sharma | 73 | 92 |
| **Tableau** | Arjun Verma | 95 | 95 |
|  | Meera Pillai | 82 | 95 |
|  | Sneha Kulkarni | 74 | 95 |

insight.
. Tara Mehta (Data Science) has the highest score overall (93).
. Arjun Verma leads Tableau with 95.
. Sia Bansal tops SQL with 92.
. Kabir Khanna leads Python with 91.
```
